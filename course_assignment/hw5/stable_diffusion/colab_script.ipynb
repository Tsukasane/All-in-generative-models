# Download Data

```python
import os
import requests

# Pretrained diffusion ckpt
url = "https://huggingface.co/runwayml/stable-diffusion-v1-5/resolve/main/v1-5-pruned-emaonly.ckpt"

save_dir = "/content/All-in-generative-models/course_assignment/hw5/stable_diffusion/models/"
os.makedirs(save_dir, exist_ok=True)

filename = os.path.join(save_dir, "v1-5-pruned-emaonly.ckpt")

response = requests.get(url, stream=True)
if response.status_code == 200:
    with open(filename, 'wb') as f:
        for chunk in response.iter_content(chunk_size=8192):
            if chunk:  # filter out keep-alive new chunks
                f.write(chunk)
    print(f"File successfully downloaded and saved to {filename}")
else:
    print("Failed to download file")

# 验证文件是否成功下载
if os.path.exists(filename):
    print(f"Checkpoint file is located at: {filename}")
else:
    print("Checkpoint file was not found")
```
